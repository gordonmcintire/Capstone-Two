```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df
```

```python
df = pd.read_csv('eplresults.csv')
df['FTR'].value_counts()
df=df.replace('H', 'Home Win')
df=df.replace('A', 'Away Win')
df=df.replace('D', 'Draw')
df2=df[['Season', 'DateTime', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR']]
```

```python
covidseason = df.loc[10044:10423]
covidseason['FTR'].value_counts()
covidseason
```

```python
sns.countplot(data=covidseason,
              x='FTR')
plt.xlabel('')
plt.ylabel('Count')
plt.title('2020/2021 Match Results')
sns.despine()
plt.show()
```

```python
sns.countplot(data=df2,
              x='FTR', order=['Home Win', 'Away Win', 'Draw'])
plt.xlabel('')
plt.ylabel('Count')
plt.title('2020/2021 Match Results')
sns.despine()
plt.show()
```
